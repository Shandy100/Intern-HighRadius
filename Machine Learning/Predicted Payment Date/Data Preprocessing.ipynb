```python
import pandas as pd
df = pd.read_csv(r'Dataset.csv')
df.head()
```

```python
#clear_date,buisness_year,posting_date,document_create_date,document_create_date.1,due_in_date,baseline_create_date
df.isnull().sum()
```

```python
df.drop('area_business',axis=1,inplace=True) #Dropping area_business since itsnull
df.head()
```

```python
df['buisness_year'] = pd.to_datetime(df['buisness_year'],format='%Y').dt.year #converting to date format
```

```python
df.head()
```

```python
df['clear_date'] = pd.to_datetime(df['clear_date'])#converting to date format
df['posting_date'] = pd.to_datetime(df['posting_date'])#converting to date format
df['document_create_date'] = pd.to_datetime(df['document_create_date'],format='%Y%m%d')#converting to date format
df['document_create_date.1'] = pd.to_datetime(df['document_create_date.1'],format='%Y%m%d')#converting to date format
df['due_in_date'] = pd.to_datetime(df['due_in_date'],format='%Y%m%d')#converting to date format
df['baseline_create_date'] = pd.to_datetime(df['baseline_create_date'],format='%Y%m%d')#converting to date format
```
